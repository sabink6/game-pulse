```python
import re
import nltk
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from gensim import corpora
from gensim.models.phrases import Phrases, Phraser
from gensim.models.coherencemodel import CoherenceModel
from gensim.models.ldamodel import LdaModel
import pyLDAvis.gensim_models as gensimvis
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import STOPWORDS
import logging

logging.getLogger("gensim").setLevel(logging.WARNING)

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load dataset
file_path = "new_dataset.csv"
df_new = pd.read_csv(file_path)

# Load stopwords and create a custom list
stop_words = set(nltk.corpus.stopwords.words('english'))
custom_stopwords = stop_words.union({'game', 'one', 'also', 'still', 'get', 'would', 'even', 'new', 'well'})

# Initialize lemmatizer
lemmatizer = nltk.stem.WordNetLemmatizer()

def preprocess_text(text):
    """Tokenizes, removes stopwords, applies lemmatization."""
    words = re.findall(r'\b\w+\b', text.lower())
    words = [lemmatizer.lemmatize(word) for word in words if word not in custom_stopwords]
    return words

# Expand heuristic categories into separate rows per category
df_new_expanded = df_new.explode('heuristic')
df_new_expanded['heuristic'] = df_new_expanded['heuristic'].str.replace(r"[\(\)']", "", regex=True).str.strip()

# Select only informative reviews
df_new_filtered = df_new_expanded[df_new_expanded['informative'] == 1]

# Define sentiment mapping based on BERT (primary) and VADER (fallback)
def classify_sentiment(row):
    if row['bert_label'] in ['Positive', 'Negative']:
        return row['bert_label'].lower()
    return row['vader_label'].lower()

df_new_filtered['sentiment'] = df_new_filtered.apply(classify_sentiment, axis=1)

# Train bigram and trigram models
sentences = df_new_filtered['clean_sentence'].dropna().tolist()
tokenized_sentences = [preprocess_text(sentence) for sentence in sentences]

bigram = Phrases(tokenized_sentences, min_count=5, threshold=10)
trigram = Phrases(bigram[tokenized_sentences], threshold=10)

bigram_mod = Phraser(bigram)
trigram_mod = Phraser(trigram)

def apply_bigrams_trigrams(tokens):
    """Applies bigram and trigram models to tokenized text."""
    return trigram_mod[bigram_mod[tokens]]

def perform_lda(sentences, num_topics=3):
    """Performs LDA topic modeling and returns the model, dictionary, and corpus."""
    tokenized_texts = [apply_bigrams_trigrams(preprocess_text(sentence)) for sentence in sentences]
    dictionary = corpora.Dictionary(tokenized_texts)
    corpus = [dictionary.doc2bow(text) for text in tokenized_texts]
    
    if len(dictionary) > 0 and len(corpus) > 0:
        lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=20, iterations=500)
        return lda_model, dictionary, corpus, tokenized_texts
    return None, None, None, None


def find_best_num_topics(sentences, topic_range=[3, 5, 7]):
    """Finds the best number of topics based on coherence score."""
    best_score = 0
    best_model = None
    
    lda_model, dictionary, corpus, tokenized_texts = perform_lda(sentences, num_topics=3)
    if not lda_model:
        return None, None, None
    
    for num_topics in topic_range:
        model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=20, iterations=500)
        coherence_model = CoherenceModel(model=model, texts=tokenized_texts, dictionary=dictionary, coherence='c_v')
        score = coherence_model.get_coherence()
        
        if score > best_score:
            best_score = score
            best_model = model
    
    return best_model, dictionary, corpus

def enhanced_wordcloud(lda_model, dictionary, sentiment):
    """Generates a word cloud for LDA topics with color-coded sentiment."""
    all_words = {}
    for i in range(lda_model.num_topics):
        topic_words = lda_model.show_topic(i, 20)
        if not topic_words:
            continue  # Skip empty topics
        for word, weight in topic_words:
            all_words[word] = weight
    
    if not all_words:
        print(f"No words available for {sentiment} word cloud.")
        return
    
    wordcloud = WordCloud(width=800, height=400, colormap='RdYlGn' if sentiment == 'positive' else 'coolwarm',
                          stopwords=STOPWORDS, background_color='black').generate_from_frequencies(all_words)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(f"Word Cloud for {sentiment.capitalize()} Reviews")
    plt.show()

def generate_insights(lda_model, dictionary, category, sentiment):
    """Generates structured insights displaying topics and keywords."""
    topics = lda_model.show_topics(num_words=10, formatted=False)
    print(f"\n{category} ({sentiment.capitalize()} Topics):")
    for topic_id, words in topics:
        keywords = ", ".join([word[0] for word in words])
        print(f"  - Topic {topic_id + 1}: {keywords}")

def extract_top_keywords(sentences, top_n=10):
    """Extracts most important keywords using TF-IDF."""
    vectorizer = TfidfVectorizer(stop_words='english', max_features=top_n)
    X = vectorizer.fit_transform(sentences)
    keywords = vectorizer.get_feature_names_out()
    print(f"Top {top_n} Keywords: {', '.join(keywords)}")

def analyze_and_visualize(category, sentiment):
    """Perform LDA, visualize topics, and generate insights."""
    subset = df_new_filtered[(df_new_filtered['heuristic'] == category) & (df_new_filtered['sentiment'] == sentiment)]
    sentences = subset['clean_sentence'].dropna().tolist()
    
    if not sentences:
        print(f"No {sentiment} reviews found for {category}.")
        return
    
    print(f"\nFinding best topics for {category} ({sentiment.capitalize()})...")
    best_model, dictionary, corpus = find_best_num_topics(sentences)
    
    if best_model:
        enhanced_wordcloud(best_model, dictionary, sentiment)
        generate_insights(best_model, dictionary, category, sentiment)
        extract_top_keywords(sentences)

# Perform LDA for each heuristic category and sentiment
distinct_categories = df_new_filtered['heuristic'].unique()
for category in distinct_categories:
    for sentiment in ['positive', 'negative']:
        analyze_and_visualize(category, sentiment)
```
