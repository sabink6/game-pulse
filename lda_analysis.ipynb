Performs LDA on positive and negative reviews for each heuristic category.
Key Features of This Script
✅ Custom Stopwords: Removes non-informative words (e.g., "game", "still", "get").
✅ Lemmatization: Converts words to their root form for consistency.
✅ Phrase Detection: Identifies bigrams ("user interface") and trigrams ("space station docking").
✅ Optimal Topic Count: Finds the best num_topics based on coherence score.
✅ WordCloud Visualization: Displays word clouds of dominant topic words.
✅ Interactive LDA Plot: Uses pyLDAvis to analyze topic distributions.

```python
import pandas as pd
import re
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from gensim import corpora
from gensim.models.phrases import Phrases, Phraser
from gensim.models.coherencemodel import CoherenceModel
from gensim.models.ldamodel import LdaModel
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load dataset (replace with actual path)
file_path = "sample_labelled.xlsx"
df = pd.read_excel(file_path, sheet_name="in")

# Filter only informative sentences
df_filtered = df[df['informative'] == 1]

# Heuristic categories to analyze
categories = ['F', 'G', 'U']

# Load stopwords and create a custom list
stop_words = set(nltk.corpus.stopwords.words('english'))
custom_stopwords = stop_words.union({'game', 'one', 'also', 'still', 'get', 'would', 'even', 'new', 'well'})

# Initialize lemmatizer
lemmatizer = nltk.stem.WordNetLemmatizer()

# Function to clean and preprocess text
def preprocess_text(text):
    """Tokenizes, removes stopwords, applies lemmatization."""
    words = re.findall(r'\b\w+\b', text.lower())  # Tokenization
    words = [lemmatizer.lemmatize(word) for word in words if word not in custom_stopwords]  # Stopword removal & lemmatization
    return words

# Train bigram and trigram models
sentences = df_filtered['clean_sentence'].dropna().tolist()
tokenized_sentences = [preprocess_text(sentence) for sentence in sentences]

bigram = Phrases(tokenized_sentences, min_count=5, threshold=10)
trigram = Phrases(bigram[tokenized_sentences], threshold=10)

bigram_mod = Phraser(bigram)
trigram_mod = Phraser(trigram)

def apply_bigrams_trigrams(tokens):
    """Applies bigram and trigram models to tokenized text."""
    return trigram_mod[bigram_mod[tokens]]

# Function to perform LDA topic modeling
def perform_lda(sentences, num_topics=3):
    """Performs LDA topic modeling and returns the model, dictionary, and corpus."""
    tokenized_texts = [apply_bigrams_trigrams(preprocess_text(sentence)) for sentence in sentences]
    dictionary = corpora.Dictionary(tokenized_texts)
    corpus = [dictionary.doc2bow(text) for text in tokenized_texts]

    if len(dictionary) > 0 and len(corpus) > 0:
        lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=20, iterations=500)
        return lda_model, dictionary, corpus
    return None, None, None

# Function to find the best number of topics
def find_best_num_topics(sentences, topic_range=[3, 5, 7]):
    """Finds the best number of topics based on coherence score."""
    tokenized_texts = [apply_bigrams_trigrams(preprocess_text(sentence)) for sentence in sentences]
    dictionary = corpora.Dictionary(tokenized_texts)
    corpus = [dictionary.doc2bow(text) for text in tokenized_texts]

    best_score = 0
    best_num_topics = topic_range[0]
    best_model = None

    for num_topics in topic_range:
        model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=20, iterations=500)
        coherence_model = CoherenceModel(model=model, texts=tokenized_texts, dictionary=dictionary, coherence='c_v')
        score = coherence_model.get_coherence()
        print(f"Coherence Score for {num_topics} topics: {score}")

        if score > best_score:
            best_score = score
            best_num_topics = num_topics
            best_model = model

    print(f"Best Number of Topics: {best_num_topics} (Score: {best_score})")
    return best_model, dictionary, corpus

# Function to extract and analyze topics
def analyze_topics(lda_model, dictionary, num_topics):
    """Extracts LDA topics and returns structured topic information."""
    topics = lda_model.show_topics(num_topics=num_topics, formatted=False)
    topic_analysis = []

    for topic_id, words in topics:
        keywords = [word[0] for word in words]
        topic_analysis.append((topic_id, keywords))

    return topic_analysis

# Function to generate structured insights
def generate_insights(category, topics, sentiment):
    """Generates insights based on identified topics."""
    print(f"\n{category}️⃣ {category} (Functionality/GamePlay/Usability)\n")

    print(f"📌 {sentiment.capitalize()} Topics (Strengths)" if sentiment == "positive" else f"⚠️ {sentiment.capitalize()} Topics (Weaknesses)")
    
    for topic_id, keywords in topics:
        print(f"  - **Topic {topic_id}**: {', '.join(keywords[:6])}")

    print("\n✅ Insights:")
    if sentiment == "positive":
        print("- Players appreciate the improvements in these areas.")
        print("- These aspects contribute to a better gameplay experience.")
    else:
        print("- Common complaints revolve around these issues.")
        print("- These areas need urgent attention from developers.")

    print("\n🛠 Suggested Improvements:")
    if sentiment == "positive":
        print("- Continue refining these areas for better user satisfaction.")
    else:
        print("- Address these concerns in future updates.")
    
    print("\n" + "=" * 50 + "\n")

# Function to visualize LDA topics using WordClouds
def visualize_wordclouds(lda_model, dictionary, num_topics):
    """Generates word clouds for LDA topics."""
    fig, axes = plt.subplots(1, num_topics, figsize=(15, 5))
    for i in range(num_topics):
        topic_words = dict(lda_model.show_topic(i, 20))  # Get top 20 words
        wordcloud = WordCloud(width=400, height=400, background_color='white').generate_from_frequencies(topic_words)
        axes[i].imshow(wordcloud, interpolation='bilinear')
        axes[i].axis("off")
        axes[i].set_title(f"Topic {i+1}")

    plt.tight_layout()
    plt.show()

# Function to visualize LDA topics interactively
def visualize_pyldavis(lda_model, corpus, dictionary):
    """Generates an interactive LDA visualization inline without requiring an external server."""
    vis = gensimvis.prepare(lda_model, corpus, dictionary)
    
    # Display inline inside Jupyter Notebook / Spyder
    import IPython.display
    IPython.display.display(pyLDAvis.display(vis))

# Dictionary to store LDA results
lda_results = {}

for category in categories:
    # Filter sentences by heuristic category
    category_df = df_filtered[df_filtered['heuristic'] == category]

    # Separate positive and negative sentences
    positive_sentences = category_df[category_df['voted_up'] == 1]['clean_sentence'].dropna().tolist()
    negative_sentences = category_df[category_df['voted_up'] == 0]['clean_sentence'].dropna().tolist()

    # Find best number of topics
    print(f"\nFinding best topics for {category} (Positive)...")
    best_model_pos, dictionary_pos, corpus_pos = find_best_num_topics(positive_sentences)
    
    print(f"\nFinding best topics for {category} (Negative)...")
    best_model_neg, dictionary_neg, corpus_neg = find_best_num_topics(negative_sentences)

    # Store results
    lda_results[category] = {
        "positive": best_model_pos,
        "negative": best_model_neg
    }

    # Visualize topics and get insights for positive feedback
    if best_model_pos:
        print(f"\nVisualizing WordCloud for {category} (Positive)...")
        visualize_wordclouds(best_model_pos, dictionary_pos, num_topics=best_model_pos.num_topics)
        print(f"\nInteractive LDA Visualization for {category} (Positive)...")
        visualize_pyldavis(best_model_pos, corpus_pos, dictionary_pos)
        pos_topics = analyze_topics(best_model_pos, dictionary_pos, num_topics=best_model_pos.num_topics)
        generate_insights(category, pos_topics, "positive")

    # Visualize topics and get insights for negative feedback
    if best_model_neg:
        print(f"\nVisualizing WordCloud for {category} (Negative)...")
        visualize_wordclouds(best_model_neg, dictionary_neg, num_topics=best_model_neg.num_topics)
        print(f"\nInteractive LDA Visualization for {category} (Negative)...")
        visualize_pyldavis(best_model_neg, corpus_neg, dictionary_neg)
        neg_topics = analyze_topics(best_model_neg, dictionary_neg, num_topics=best_model_neg.num_topics)
        generate_insights(category, neg_topics, "negative")
```
